```python
class InformationItem:
    def __init__(self, title, content):
        self.title = title
        self.content = content

class KnowledgeBase:
    def __init__(self):
        self.topics = {}

    def add_topic(self, topic):
        if topic not in self.topics:
            self.topics[topic] = []
            print(f"Topic '{topic}' added successfully.")
        else:
            print("Topic already exists.")

    def add_information(self, topic, title, content):
        if topic in self.topics:
            self.topics[topic].append(InformationItem(title, content))
            print(f"Information added to topic '{topic}'.")
        else:
            print("Topic not found.")

    def view_topic(self, topic):
        if topic in self.topics:
            print(f"Information for topic '{topic}':")
            for info_item in self.topics[topic]:
                print(f"Title: {info_item.title}")
                print(f"Content: {info_item.content}")
                print()
        else:
            print("Topic not found.")

# Sample Usage
if __name__ == "__main__":
    knowledge_base = KnowledgeBase()

    while True:
        print("\n1. Add Topic\n2. Add Information\n3. View Topic\n4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            topic = input("Enter Topic Name: ")
            knowledge_base.add_topic(topic)
        elif choice == '2':
            topic = input("Enter Topic Name: ")
            title = input("Enter Information Title: ")
            content = input("Enter Information Content: ")
            knowledge_base.add_information(topic, title, content)
        elif choice == '3':
            topic = input("Enter Topic Name: ")
            knowledge_base.view_topic(topic)
        elif choice == '4':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please try again.")
```
