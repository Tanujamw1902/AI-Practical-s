```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_graph(graph, color_map):
    colors = list(color_map.values())
    nx.draw(graph, with_labels=True, node_color=colors, node_size=1000, font_size=12)
    plt.show()

def graph_coloring(graph, num_colors, colors):
    color_map = {}
    for node in graph.nodes():
        available_colors = set(colors[:num_colors])  # Limit available colors based on num_colors
        for neighbor in graph.neighbors(node):
            if neighbor in color_map:
                neighbor_color = color_map[neighbor]
                if neighbor_color in available_colors:
                    available_colors.remove(neighbor_color)
        color_map[node] = next(iter(available_colors))
    return color_map

def main():
    n = int(input("Enter the number of nodes in the graph: "))
    edges = int(input("Enter the number of edges: "))
    graph = nx.gnm_random_graph(n, edges)

    print("\nGraph:")
    nx.draw(graph, with_labels=True)
    plt.show()

    num_colors = int(input("Enter the number of colors: "))
    print("Enter colors (separated by space):")
    colors = input().split()

    color_map = graph_coloring(graph, num_colors, colors)
    print("\nGraph Coloring:")
    print(color_map)

    print("\nGraph with Colors:")
    plot_graph(graph, color_map)

if __name__ == "__main__":
    main()
```
