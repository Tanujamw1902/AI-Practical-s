```python
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id
        self.performance = None

    def evaluate_performance(self, performance):
        self.performance = performance

class PerformanceEvaluation:
    def __init__(self):
        self.employees = {}

    def add_employee(self, name, emp_id):
        if emp_id not in self.employees:
            self.employees[emp_id] = Employee(name, emp_id)
            print(f"Employee {name} added successfully.")
        else:
            print("Employee ID already exists.")

    def evaluate_employee(self, emp_id, performance):
        if emp_id in self.employees:
            self.employees[emp_id].evaluate_performance(performance)
            print(f"Performance evaluated for employee {emp_id}.")
        else:
            print("Employee ID not found.")

    def view_report(self):
        print("Performance Report:")
        for emp_id, employee in self.employees.items():
            if employee.performance:
                print(f"Employee ID: {emp_id}, Name: {employee.name}, Performance: {employee.performance}")
            else:
                print(f"Employee ID: {emp_id}, Name: {employee.name}, Performance: Not evaluated yet")

# Sample Usage
if __name__ == "__main__":
    performance_eval = PerformanceEvaluation()

    while True:
        print("\n1. Add Employee\n2. Evaluate Employee Performance\n3. View Performance Report\n4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter Employee Name: ")
            emp_id = input("Enter Employee ID: ")
            performance_eval.add_employee(name, emp_id)
        elif choice == '2':
            emp_id = input("Enter Employee ID: ")
            performance = input("Enter Employee Performance: ")
            performance_eval.evaluate_employee(emp_id, performance)
        elif choice == '3':
            performance_eval.view_report()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please try again.")
```
