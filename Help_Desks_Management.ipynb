```python
class Ticket:
    def __init__(self, ticket_id, title, description, status="Open"):
        self.ticket_id = ticket_id
        self.title = title
        self.description = description
        self.status = status

class HelpDesk:
    def __init__(self):
        self.tickets = {}
        self.ticket_counter = 1

    def log_ticket(self, title, description):
        ticket_id = self.ticket_counter
        self.tickets[ticket_id] = Ticket(ticket_id, title, description)
        self.ticket_counter += 1
        print(f"Ticket logged successfully with ID: {ticket_id}")

    def update_ticket_status(self, ticket_id, new_status):
        if ticket_id in self.tickets:
            self.tickets[ticket_id].status = new_status
            print(f"Ticket {ticket_id} status updated to {new_status}.")
        else:
            print("Ticket not found.")

    def view_ticket(self, ticket_id):
        if ticket_id in self.tickets:
            ticket = self.tickets[ticket_id]
            print(f"Ticket ID: {ticket.ticket_id}")
            print(f"Title: {ticket.title}")
            print(f"Description: {ticket.description}")
            print(f"Status: {ticket.status}")
        else:
            print("Ticket not found.")

    def view_all_tickets(self):
        if self.tickets:
            print("All Tickets:")
            for ticket_id, ticket in self.tickets.items():
                print(f"Ticket ID: {ticket_id}, Title: {ticket.title}, Status: {ticket.status}")
        else:
            print("No tickets logged yet.")

# Sample Usage
if __name__ == "__main__":
    help_desk = HelpDesk()

    while True:
        print("\nHelp Desk Management System")
        print("1. Log Ticket\n2. Update Ticket Status\n3. View Ticket\n4. View All Tickets\n5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            title = input("Enter Ticket Title: ")
            description = input("Enter Ticket Description: ")
            help_desk.log_ticket(title, description)
        elif choice == '2':
            ticket_id = int(input("Enter Ticket ID: "))
            new_status = input("Enter New Status: ")
            help_desk.update_ticket_status(ticket_id, new_status)
        elif choice == '3':
            ticket_id = int(input("Enter Ticket ID: "))
            help_desk.view_ticket(ticket_id)
        elif choice == '4':
            help_desk.view_all_tickets()
        elif choice == '5':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please try again.")
```
