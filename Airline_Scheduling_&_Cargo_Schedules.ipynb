```python
class Flight:
    def __init__(self, flight_id, airline, destination, departure_time, status="Scheduled"):
        self.flight_id = flight_id
        self.airline = airline
        self.destination = destination
        self.departure_time = departure_time
        self.status = status

class AirlineScheduler:
    def __init__(self):
        self.flights = {}
        self.flight_counter = 1

    def schedule_flight(self, airline, destination, departure_time):
        flight_id = self.flight_counter
        self.flights[flight_id] = Flight(flight_id, airline, destination, departure_time)
        self.flight_counter += 1
        print(f"Flight scheduled successfully with ID: {flight_id}")

    def update_flight_status(self, flight_id, new_status):
        if flight_id in self.flights:
            self.flights[flight_id].status = new_status
            print(f"Flight {flight_id} status updated to {new_status}.")
        else:
            print("Flight not found.")

    def view_flight(self, flight_id):
        if flight_id in self.flights:
            flight = self.flights[flight_id]
            print(f"Flight ID: {flight.flight_id}")
            print(f"Airline: {flight.airline}")
            print(f"Destination: {flight.destination}")
            print(f"Departure Time: {flight.departure_time}")
            print(f"Status: {flight.status}")
        else:
            print("Flight not found.")

    def view_all_flights(self):
        if self.flights:
            print("All Flights:")
            for flight_id, flight in self.flights.items():
                print(f"Flight ID: {flight_id}, Airline: {flight.airline}, Destination: {flight.destination}, Status: {flight.status}")
        else:
            print("No flights scheduled yet.")

# Sample Usage
if __name__ == "__main__":
    airline_scheduler = AirlineScheduler()

    while True:
        print("\nAirline Scheduling & Cargo Schedules")
        print("1. Schedule Flight\n2. Update Flight Status\n3. View Flight\n4. View All Flights\n5. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            airline = input("Enter Airline Name: ")
            destination = input("Enter Destination: ")
            departure_time = input("Enter Departure Time: ")
            airline_scheduler.schedule_flight(airline, destination, departure_time)
        elif choice == '2':
            flight_id = int(input("Enter Flight ID: "))
            new_status = input("Enter New Status: ")
            airline_scheduler.update_flight_status(flight_id, new_status)
        elif choice == '3':
            flight_id = int(input("Enter Flight ID: "))
            airline_scheduler.view_flight(flight_id)
        elif choice == '4':
            airline_scheduler.view_all_flights()
        elif choice == '5':
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please try again.")
```
